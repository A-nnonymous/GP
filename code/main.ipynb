```python
import folium

print(folium.__version__)
```

```python
import pandas as pd
DataPath="../data/csse_covid_19_data/"
SeriesDataPath="csse_covid_19_time_series/"
DailyPath="csse_covid_19_daily_reports_us/"
f=open(DataPath+DailyPath+"03-12-2021.csv",'r')
data=pd.read_csv(f)
data.head()
```

```python
import numpy as np
data = data.fillna(1)
lat=data['Lat'].values.tolist()
long=data['Long_'].values.tolist()
confirmed=data['Confirmed'].values
confirmed=np.log(confirmed+1)
confirmed=confirmed.tolist()
```

```python
heatdata=list(zip(lat,long,confirmed))
```

```python
data.describe()
```

```python
from folium.plugins import HeatMap
m = folium.Map([36.840089 ,-85.206614], tiles='stamentoner', zoom_start=4)
# HeatMap
HeatMap(heatdata,radius=35).add_to(m)
# m.save(os.path.join('results', 'Heatmap.html'))
m
```
